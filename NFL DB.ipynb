### QB College Boxplots Passing

```python
%matplotlib inline

from glob import glob
from pandas import read_csv
import pandas as pb
from datetime import datetime
from collections import defaultdict
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
xerr = []
ci_data = pb.read_csv("college_ci_passing_data.csv",header=0)
ci_data["BS_PlotMin"] = abs(ci_data.BS_Min - ci_data.Mean)
ci_data["BS_PlotMax"] = abs(ci_data.BS_Max - ci_data.Mean)

for i,group in ci_data.groupby("Team"):
    xerr.append((group.BS_PlotMin.max(),group.BS_PlotMax.max()))

plt.figure(figsize=(8,20))

plt.barh(range(len(ci_data)),ci_data.Mean,xerr=xerr)
plt.yticks(np.arange(len(ci_data)),ci_data.Team.values)
plt.savefig("team-passing-box.png",bbox_inches="tight")
```
