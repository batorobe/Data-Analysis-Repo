```python
#most famous ships/size comparison, height comparison height on y-axis, names on x-axis
#talk about crew capacities compare crew of death star to cities in US

import swapi
from collections import defaultdict

starships = swapi.get_all("starships")
vehicles = swapi.get_all("vehicles")
planets = swapi.get_all("planets")
species = swapi.get_all("species")
people = swapi.get_all("people")
```

```python
import string
import re
exclude = set(string.punctuation)
non_dec = re.compile(r'[^\d.]+')
ship_array = []
vehicle_array = []
planet_array = []
person_array = []

for ship in starships.iter():
    ship_length = ''.join(ch for ch in ship.length if ch not in exclude)
    ship_array.append((ship.name,float(ship_length)))
for vehicle in vehicles.iter():
    if(vehicle.length!='unknown'):
        vehicle_array.append((vehicle.name, float(vehicle.length)))
for planet in planets.iter():
    if((planet.diameter!='unknown') and (planet.diameter>1)):
        planet_array.append((planet.name, float(planet.diameter)))
for person in people.iter():
    if(person.height!='unknown'):
        person_array.append((person.name, float(person.height)))
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from math import log10

vehicle_df = pd.DataFrame(vehicle_array,columns=["Vehicle_Name","Size_m"])
vehicle_df["Size_m"]=vehicle_df["Size_m"].apply(lambda x: (x*3.28084))
vehicle_df = vehicle_df.sort('Size_m',ascending=True)
vehicle_df.Size_m.apply(log10).plot(kind='bar', figsize=(12,8))
plt.xticks(np.arange(0,38,1),vehicle_df.Vehicle_Name.values)
plt.yticks(np.arange(0,4,1),[0,10,100,1000])
plt.title("SWAPI Vehicle Length")
plt.ylabel("Log Scaled Length of Vehicle (ft)")
#plt.yscale('log')
plt.savefig("swapi-vehicle-len.png",bbox_inches="tight")
;
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

planet_df = pd.DataFrame(planet_array,columns=["Planet_Name","Size_m"])

planet_df["Size_m"]=planet_df["Size_m"].apply(lambda x: (x*3.28084))
planet_df = planet_df.sort('Size_m',ascending=True)
planet_df = planet_df[6:44]
planet_df.Size_m.plot(kind='bar', figsize=(12,8))
plt.xticks(np.arange(0,38,1),planet_df.Planet_Name.values)
plt.title("SWAPI Planet Diameter")
plt.ylabel("Diameter (ft)")
plt.savefig("swapi-planet-dia.png",bbox_inches="tight")
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import *


ship_df = pd.DataFrame(ship_array,columns=["Ship_Name","Size_m"])
ship_df["Size_m"]=ship_df["Size_m"].apply(lambda x: (x*3.28084))

ship_df = ship_df.sort('Size_m',ascending=True)
ship_df.Size_m.apply(log10).plot(kind='bar', figsize=(12,8))
plt.xticks(np.arange(0,36,1),ship_df.Ship_Name.values)
plt.yticks(np.arange(0,7,1),[0,10,100,1000,10000,100000,1000000])
plt.ylabel("Log Scaled Width (ft)")
plt.title("SWAPI Ship Width")
plt.savefig("swapi-ship-width.png",bbox_inches="tight")
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


person_df = pd.DataFrame(person_array,columns=["Person_Name","Height_m"])
person_df["Height_m"]=person_df["Height_m"].apply(lambda x: (x*0.028084))
person_df = person_df.sort('Height_m',ascending=True)
person_df.Height_m.plot(kind='bar', figsize=(21,12))
plt.xticks(np.arange(0,81,1),person_df.Person_Name.values)
plt.title("SWAPI Person Height")
plt.ylabel("Height (ft)")
plt.savefig("swapi-person-hei.png",bbox_inches="tight")
```
