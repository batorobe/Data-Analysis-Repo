### QB College Boxplots Passing

### SCIPY Ranksums Team Rushing 

```python
from glob import glob
from scipy import stats
from pandas import read_csv
import pandas as pb
from datetime import datetime
from collections import defaultdict
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
pvalues = defaultdict(tuple)
labels = []
nfl_passing_data = pb.read_csv("nfl_passing_qb_stats.csv",
                          skiprows=[0], names=["full_name","college","team","passing_yds","rushing_yds"])

for team1,group1 in nfl_passing_data.groupby("team"):
    labels.append(team1)
    for team2,group2 in nfl_passing_data.groupby("team"):
        pvalues[team1,team2] +=stats.ranksums(group1.rushing_yds,group2.rushing_yds)
```

```python
import itertools
 
x_label = defaultdict(str)
y_label = defaultdict(str)
heatmap_val = []
num=0
count = 0
 
for i in range(0,32):
    heatmap_val.append([])
for team1,num in zip(sorted(pvalues.keys()),range(0,1024)):
    heatmap_val[num%32].append(pvalues[team1][1])
plt.figure(figsize=(30,30))
plt.imshow(heatmap_val,cmap="Blues",interpolation="none")
plt.colorbar(shrink=0.75)
plt.yticks()
#plt.title("Percentage of Reddit Submissions Scoring > 100 - August 2014")
plt.grid(False)
plt.yticks(np.arange(0,32,1),labels)
plt.xticks(np.arange(0,32,1),labels)
plt.savefig("team-passing-pvalue-heatmap.png",bbox_inches="tight")
;
```
