```python
from glob import glob
from pandas import read_csv
import pandas as pd
from datetime import datetime
from collections import defaultdict

year = 1922
filenames = []
nline = ""
for item_object in glob("*.csv"):
    if (item_object!= "total_data.csv"):
        filenames.append(item_object)
        
with open("total_data.csv", "wb") as out_file:
    out_file.write("Year,Week,Day,Date,box,Winner/tie,@,Loser/tie,PtsW,PtsL,YdsW,TOW,YdsL,TOL")
    for fileName in filenames:
        reddit_data = pd.read_csv(fileName,header=0)
        for line in reddit_data.values:
            out_file.write("\n"+str(year)+","+','.join([str(x) for x in line]) )
        year +=1     
```
